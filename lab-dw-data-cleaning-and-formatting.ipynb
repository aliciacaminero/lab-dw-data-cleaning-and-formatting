```python
# 1. Identificar columnas con valores nulos
valores_nulos = company_df.isnull().sum()
print("Valores nulos por columna:")
print(valores_nulos[valores_nulos > 0])  # Mostrar solo columnas con valores nulos
```

```python
# 2. Decidir una estrategia para manejar los valores nulos
# Supongamos que elegimos rellenar valores nulos numéricos con la mediana
# y valores nulos categóricos con la moda.
```

```python
# 3. Rellenar valores nulos
for column in company_df.columns:
    if company_df[column].dtype == 'object':  # Categórica
        moda = company_df[column].mode()[0]  # Tomar la moda
        company_df[column].fillna(moda, inplace=True)
    else:  # Numérica
        mediana = company_df[column].median()  # Tomar la mediana
        company_df[column].fillna(mediana, inplace=True)
```

```python
# 4. Verificar que la estrategia ha manejado con éxito los valores nulos
valores_nulos_actualizados = company_df.isnull().sum()
print("\nValores nulos por columna después de la imputación:")
print(valores_nulos_actualizados[valores_nulos_actualizados > 0])  # Mostrar solo columnas con valores nulos
```

```python
# 5. Convertir todas las variables numéricas a enteros
# Asegúrate de que no hay valores nulos antes de convertir
company_df = company_df.fillna(0)  # Llenar nulos con 0 antes de la conversión (si aplica)
numerical_cols = company_df.select_dtypes(include=['float64']).columns
company_df[numerical_cols] = company_df[numerical_cols].astype(int)
```

```python
# 6. Mostrar los tipos de datos finales
print("\nTipos de datos finales:")
print(company_df.dtypes)
```

Put all the data cleaning and formatting steps into functions, and create a main function that performs all the cleaning and formatting.

```python
# Script para crear data_cleaning.py

data_cleaning_code = """
import pandas as pd

def load_data(filepath):
    \"\"\"Carga el DataFrame desde un archivo CSV.\"\"\"
    return pd.read_csv(filepath)

def identify_nulls(df):
    \"\"\"Identifica y muestra el número de valores nulos por columna.\"\"\"
    return df.isnull().sum()

def handle_nulls(df):
    \"\"\"Maneja los valores nulos en el DataFrame.\"\"\"
    for column in df.columns:
        if df[column].dtype == 'object':  # Categórica
            moda = df[column].mode()[0]  # Tomar la moda
            df[column].fillna(moda, inplace=True)
        else:  # Numérica
            mediana = df[column].median()  # Tomar la mediana
            df[column].fillna(mediana, inplace=True)
    return df

def handle_duplicates(df):
    \"\"\"Elimina filas duplicadas y devuelve el DataFrame depurado.\"\"\"
    df = df.drop_duplicates(keep='first')
    return df

def normalize_column_names(df):
    \"\"\"Normaliza los nombres de las columnas en el DataFrame.\"\"\"
    # Renombrar columnas específicas
    df.rename(columns={'GENDER': 'Gender', 'ST': 'State'}, inplace=True)
    df.columns = df.columns.str.lower()
    df.columns = df.columns.str.replace(' ', '_')
    return df

def convert_data_types(df):
    \"\"\"Convierte tipos de datos en el DataFrame.\"\"\"
    df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'], errors='coerce')
    df['number_of_open_complaints'] = (
        df['number_of_open_complaints']
        .str.split('/')
        .str[1]
        .astype(float)
        .fillna(0)
        .astype(int)
    )
    numerical_cols = df.select_dtypes(include=['float64']).columns
    df[numerical_cols] = df[numerical_cols].fillna(0).astype(int)
    return df

def save_data(df, filepath):
    \"\"\"Guarda el DataFrame en un archivo CSV.\"\"\"
    df.to_csv(filepath, index=False)
"""

# Guardar el código en un archivo data_cleaning.py
with open("data_cleaning.py", "w") as file:
    file.write(data_cleaning_code)

print("El archivo data_cleaning.py ha sido creado.")

```
